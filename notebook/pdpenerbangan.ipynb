```python
# === 2. Pra-pemrosesan Data ===

# Konversi kolom Durasi_Penerbangan (format "jam:menit") ke menit
def convert_to_minutes(duration):
    try:
        h, m = map(int, duration.split(":"))
        return h * 60 + m
    except:
        return None

df_final["Durasi_Penerbangan_menit"] = df_final["Durasi_Penerbangan"].apply(convert_to_minutes)
```

```python
# Pilih fitur untuk model
features = [
    "Kota_tujuan",
    "Deskripsi_tujuan",
    "Suhu_Celcius_tujuan",
    "Kelembaban_%_tujuan",
    "Tekanan_hPa_tujuan",
    "Kecepatan_Angin_m/s_tujuan"
]

target = "Durasi_Penerbangan_menit"
```

```python
# Encode fitur kategorikal (Label Encoding)
le_dict = {}
for col in ["Kota_tujuan", "Deskripsi_tujuan"]:
    le = LabelEncoder()
    df_final[col] = le.fit_transform(df_final[col])
    le_dict[col] = le
```

```python
# Buang baris yang tidak punya target
df_final = df_final.dropna(subset=[target])

X = df_final[features]
y = df_final[target]
```

```python
# =====================================
# 7 Distribusi parameter untuk tuning
# =====================================
from scipy.stats import randint

param_dist = {
    "n_estimators": randint(100, 500),        # jumlah pohon
    "max_depth": [None, 10, 20, 30, 40, 50],  # kedalaman pohon
    "min_samples_split": randint(2, 10),      # minimal split
    "min_samples_leaf": randint(1, 5),        # minimal daun
    "max_features": ["auto", "sqrt", "log2"]  # fitur maksimum per split
}
```

```python
# =====================================
# 8 Fungsi tuning dan evaluasi
# =====================================
def tuning_dan_evaluasi(X, y, test_size):
    # Split dataset
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42
    )

    # Model dasar
    rf = RandomForestRegressor(random_state=42)
    # Randomized Search
    from sklearn.model_selection import RandomizedSearchCV
    random_search = RandomizedSearchCV(
        estimator=rf,
        param_distributions=param_dist,
        n_iter=30,          # jumlah percobaan
        cv=5,
        n_jobs=-1,
        random_state=42,
        scoring="r2",
        verbose=0
    )

    random_search.fit(X_train, y_train)

    # Model terbaik
    best_model = random_search.best_estimator_


    # Prediksi
    y_pred = best_model.predict(X_test)

    # Evaluasi
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    return {
        "split": f"{int((1-test_size)*100)}:{int(test_size*100)}",
        "best_params": random_search.best_params_,
        "cv_score": random_search.best_score_,
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2,
        "model": best_model,
        "X_test": X_test, # Return X_test
        "y_test": y_test  # Return y_test
    }
```

```python
# =====================================
# 9 Jalankan untuk 3 skenario
# =====================================
results = []
for ratio in [0.2, 0.3, 0.4]:  # 80:20, 70:30, 60:40
    res = tuning_dan_evaluasi(X, y, ratio)
    results.append(res)
```

```python
# =====================================
# 10 Tampilkan hasil
# =====================================
for res in results:
    print(f"\n Split {res['split']}")
    print(f"   Best Params: {res['best_params']}")
    print(f"   CV R Score = {res['cv_score']:.3f}")
    print(f"   MAE  = {res['MAE']:.2f}")
    print(f"   RMSE = {res['RMSE']:.2f}")
    print(f"   R   = {res['R2']:.3f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Ambil hasil evaluasi
splits = [res["split"] for res in results]
mae_vals = [res["MAE"] for res in results]
rmse_vals = [res["RMSE"] for res in results]
r2_vals = [res["R2"] for res in results]

# Posisi bar
y = np.arange(len(splits))
height = 0.25

# Plot
plt.figure(figsize=(10,6))

plt.barh(y - height, mae_vals, height, label="MAE")
plt.barh(y, rmse_vals, height, label="RMSE")
plt.barh(y + height, r2_vals, height, label="R")

# Label dan judul
plt.yticks(y, splits)
plt.ylabel("Skenario Split Data")
plt.xlabel("Nilai Evaluasi")
plt.title("Perbandingan Evaluasi Model Random Forest (Horizontal Bar Chart)")
plt.legend()

# Tampilkan angka di samping bar
for i, v in enumerate(mae_vals):
    plt.text(v + 0.02, y[i] - height, f"{v:.2f}", va="center", fontsize=8)
for i, v in enumerate(rmse_vals):
    plt.text(v + 0.02, y[i], f"{v:.2f}", va="center", fontsize=8)
for i, v in enumerate(r2_vals):
    plt.text(v + 0.02, y[i] + height, f"{v:.2f}", va="center", fontsize=8)

plt.tight_layout()
plt.show()
```
